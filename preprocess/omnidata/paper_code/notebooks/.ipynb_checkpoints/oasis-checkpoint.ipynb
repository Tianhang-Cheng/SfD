```python
from PIL import Image, ImageFilter 
import numpy as np

image = Image.open('/home/ainaz/Desktop/point_8_view_0_domain_rgb.png') 
depth = np.array(Image.open('/home/ainaz/Desktop/point_8_view_0_domain_depth_zbuffer.png'))

h, w = depth.shape[0], depth.shape[1]

k=4
for i in range(h // 4):
    for j in range(w // 4):
        smol_image = image.crop((j*k, i*k, (j+1)*k, (i+1)*k)) 
        depth_mean = depth[i*k:(i+1)*k, j*k:(j+1)*k].mean()
        r = 10 * (depth_mean / 2**16)
        print(r)
        blurred_image = smol_image.filter(ImageFilter.GaussianBlur(radius=r)) 
        image.paste(blurred_image, (j*k,i*k)) 

# Cropping the image 
# smol_image = image.crop((100, 0, 200, 500)) 

# # Blurring on the cropped image 
# blurred_image = smol_image.filter(ImageFilter.GaussianBlur(radius=10)) 

# # Pasting the blurred image on the original image 
# image.paste(blurred_image, (100,0)) 

# Displaying the image 
image.save('/home/ainaz/Desktop/output.png') 
```

```python
def blur(i, j):
    smol_image = image.crop((j*k, i*k, (j+1)*k, (i+1)*k)) 
    depth_mean = depth[i*k:(i+1)*k, j*k:(j+1)*k].mean()
    r = 20 ** ((depth_mean - depth_min) / 2**16)
    blurred_image = smol_image.filter(ImageFilter.GaussianBlur(radius=r)) 
#     blurred_image = smol_image.filter(ImageFilter.Kernel(size=3, scale=1)) 
    
    image.paste(blurred_image, (j*k,i*k)) 
    
image = Image.open('/home/ainaz/Desktop/point_8_view_0_domain_rgb.png') 
depth = np.array(Image.open('/home/ainaz/Desktop/point_8_view_0_domain_depth_zbuffer.png'))

h, w = depth.shape[0], depth.shape[1]
depth_min = depth.min()
k=16

i_list, j_list = [], []
for i in range(h // k):
    for j in range(w // k):
        i_list.append(i)
        j_list.append(j)
        
blur_vectorized = np.vectorize(blur)
blur_vectorized(i=i_list, j=j_list)
# image = image.filter(ImageFilter.EDGE_ENHANCE) 
image = image.filter(ImageFilter.SMOOTH)

image.save('/home/ainaz/Desktop/output2.png') 
```
